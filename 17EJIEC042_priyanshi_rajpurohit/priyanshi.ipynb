```python
im = Image.open("rose.jpeg")
im=np.array(im) #conv image to array
row, col, color = im.shape 

for i in range(row):
    for j in range(col):
        #print(im[i][j])
        if im[i][j][0] > 250:   #0=red,1=green,2=blue
            im[i][j] = ( im[i,j,0] //3 + im[i,j,1] //3 + im[i,j,2]//3)
        elif im [i][j][1] < 35  or im[i][j][2] <35:
            im[i][j] =  ( im[i,j,0] //3 + im[i,j,1] //3 + im[i,j,2]//3)
# if red>150 and b&g<50 the we will get red colour.... so we blackes all colours except red
        else :
            pass
            
im2=Image.fromarray(im)  #conv array to image 
im2.save("only_red.jpg")

plt.imshow(im2)
plt.show()
```
