```python
from PIL import Image
```

```python
pic = Image.open("horse.jpg")
```

```python
pic.size
```

```python
x, y = pic.size
x //= 2
y //= 2
smaller = pic.resize((x, y))
```

```python
smaller
```

```python
from PIL import ImageFilter

edges = smaller.filter(ImageFilter.FIND_EDGES)
```

```python
edges
```

```python
bands = edges.split()
```

```python
bands[0]
```

```python
outline = bands[0].point(lambda x: 255 if x<100 else 0)
```

```python
outline
```

```python
outline.crop((10, 10, 100, 100))
```

```python
outline.save("horse.pdf")
```
